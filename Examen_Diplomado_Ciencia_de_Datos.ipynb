17. ¿Cuál es el mejor modelo de clasificación para NLP y por qué?

**R.**
El mejor modelo depende de ciertos factores de la tarea a realizarse, como su complejidad, el tamaño de los datos o los recursos computacionales disponibles. Los modelos más utilizados son:

a)	Naive Bayes: es simple y puede ser entrenado rápida y eficientemente, además de ser efectivo con bases de datos pequeñas. Sin embargo, tiene una capacidad de aprendizaje un tanto limitada.

b)	SVM: son conocidas por ser altamente precisas y considerablemente inmunes al ruido, aunque suelen tener un alto coste computacional.

c)	Regresión logística (Logistic Regression): es un modelo simple, interpretable y eficiente, además de funcionar bien para clasificaciones binarias. Sin embargo, al igual que Naive Bayes, su capacidad de aprendizaje es limitada debido a su naturaleza lineal.

d)	Bosques aleatorios (Random Forests): son bastante precisos y pueden manejar bases de datos grandes y complejas de manera efectiva, aunque tardan más tiempo en ser entrenados que otros modelos más simples.

```python
analyzer=SentimentIntensityAnalyzer()
for text in df[3]: #Con esto podemos identificar si los comentarios son negativos o positivos
  scores=analyzer.polarity_scores(text)
  print(text)
  print(scores)
```

```python
score_list=list()
for comment in df[3]:
  score=analyzer.polarity_scores(comment)
  score_list.append(score['compound'])
df['tarjet']=pd.Series(score_list)
df.head()
```

```python
df.loc[df.SS>=0.25, 'Sentiment']= 'Positive' #A partir de 0.25 el valor es positivo
df.loc[(df.SS>=0.25)&(df.SS>=-0.25), 'Sentiment']= 'Neutral'
df.loc[df.SS<=-0.25, 'Sentiment']='Negative'
df.head()
```
